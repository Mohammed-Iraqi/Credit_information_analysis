```python
# Assignment Title: Credit information analysis
# Objectives:
# 1. Analyze things close to real data
# 2. Allow yourself to solve dataset issues
# -----------------------------------------
#  [Problem 1] Understanding the content of the competition 
#  What kind of company is Home Credit?
#  Home Credit company is a company that providing a positive and safe borrowing experience to people who want loans.
#  What is expected in this competition?
#  Based on the collected data from the client it is expected that the contester to predict their clients' repayment abilities using Machine Learning methods
#  What are the benefits companies can gain by predicting this?
#  Based on the results of the Machine Learning methods the companies will decide to accept or refuse the loan requested based on the repayment abilities of the clients.
import numpy as np
import pandas as pd
# Load Train Application CSV data file
print('------------------------------- Problem #1 ------------------------------------')
df_train = pd.read_csv("application_train.csv")
print(df_train)


#  [Problem 2] Understanding the overview of data 
# -----------------------------------------------
print('------------------------------- Problem #2 ------------------------------------')
print(df_train.shape)
print(df_train.head())
print(df_train.describe())
import missingno as msno
import seaborn as sns
import matplotlib.pyplot as plt
# Check the percentage of missing values.

def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
print(missing_data(df_train).head(70))

# Bar
msno.bar(df_train.sample(307511))

numerical_feats = df_train.dtypes[df_train.dtypes != "object"].index
print("Number of Numerical features: ", len(numerical_feats))
print(df_train[numerical_feats].columns)
print("-"*80)
categorical_feats = df_train.dtypes[df_train.dtypes == "object"].index
print("Number of Categorical features: ", len(categorical_feats))
print(df_train[categorical_feats].columns)

# There is 122 of explanatory variables
# Number of Numerical features:  106
# Number of Categorical features:  16
# There is partial missing data in 67 explanatory varaibles
# The remianing 55 explanatory varaibles has no data

#  [Problem 3] Defining issues 
# ----------------------------
# Please set multiple issues / questions of your own based on the overview of the data.
#
# What is the most imprtant explanatory varaible affecting Home Credit Decision?
# What is the least important explanatory varaible affecting Home Credit Decision?
# Why are there so much explanatory varaibles without data?
# What are the correlation among the explanatory varaible? 


#  [Problem 4] Data exploration 
# -----------------------------
# We will proceed with the data search to solve assignment and questions set in Problem 3. Create at least 5 tables and graphs to use in your search, along with your thoughts.
# If you have any new questions during the exploration process, please add them to the answer to Problem 3.

def plot_stats(feature,label_rotation=False,horizontal_layout=True):
    temp = df_train[feature].value_counts()
    df1 = pd.DataFrame({feature: temp.index,'Number of contracts': temp.values})

    # Calculate the percentage of target=1 per category value
    cat_perc = df_train[[feature, 'TARGET']].groupby([feature],as_index=False).mean()
    cat_perc.sort_values(by='TARGET', ascending=False, inplace=True)
    
    if(horizontal_layout):
        fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))
    else:
        fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(12,14))
    sns.set_color_codes("pastel")
    s = sns.barplot(ax=ax1, x = feature, y="Number of contracts",data=df1)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    
    s = sns.barplot(ax=ax2, x = feature, y='TARGET', order=cat_perc[feature], data=cat_perc)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    plt.ylabel('Percent of target with value 1 [%]', fontsize=10)
    plt.tick_params(axis='both', which='major', labelsize=10)

    plt.show();
    
def plot_distribution(var):
    
    i = 0
    t1 = df_train.loc[df_train['TARGET'] != 0]
    t0 = df_train.loc[df_train['TARGET'] == 0]

    sns.set_style('whitegrid')
    plt.figure()
    fig, ax = plt.subplots(2,2,figsize=(12,12))

    for feature in var:
        i += 1
        plt.subplot(2,2,i)
        sns.kdeplot(t1[feature], bw=0.5,label="TARGET = 1")
        sns.kdeplot(t0[feature], bw=0.5,label="TARGET = 0")
        plt.ylabel('Density plot', fontsize=12)
        plt.xlabel(feature, fontsize=12)
        locs, labels = plt.xticks()
        plt.tick_params(axis='both', which='major', labelsize=12)
    plt.show();

plot_stats('CODE_GENDER')
print('The number of female clients is almost double the number of male clients. Looking to the percent of defaulted credits, males have a higher chance of not returning their loans (~10%), comparing with women (~7%).')
plot_stats('NAME_CONTRACT_TYPE')
print('Contract type Revolving loans are just a small fraction (10%) from the total number of loans; in the same time, a larger amount of Revolving loans, comparing with their frequency, are not repaid.')
plot_stats('FLAG_OWN_CAR')

plot_stats('FLAG_OWN_REALTY')
my_string = ("The clients that owns a car are almost a half of the ones that doesn't own one. The clients that owns a \n"
            "car are less likely to not repay a car that the ones that own. Both categories have not-repayment rates around 8%. \n"
            "The clients that owns real estate are more than double of the ones that doesn't own. Both categories (owning real \n"
            "estate or not owning) have not-repayment rates less than 8%.")
print(my_string)

plot_stats('NAME_FAMILY_STATUS',True, True)
my_string =(" Most of clients are married, followed by Single/not married and civil marriage. \n"
            "In terms of percentage of not repayment of loan, Civil marriage has the highest percent \n"
            "of not repayment (10%), with Widow the lowest (exception being Unknown).")
print(my_string)

plot_stats('CNT_CHILDREN')
my_string = ("Most of the clients taking a loan have no children. The number of loans associated with the clients\n"
             "with one children are 4 times smaller, the number of loans associated with the clients with two children\n"
             "are 8 times smaller; clients with 3, 4 or more children are much more rare.\n"
             "As for repayment, clients with no children, 1, 2, 3, and 5 children have percents of \n"
             "no repayment around the average (10%). The clients with 4 and 6 children are above average\n"
             "in terms of percent of not paid back loans (over 25% for families with 6 children).\n"
             "As for clients with 9 or 11 children, the percent of loans not repaid is 100%.")
print(my_string)

plot_stats('CNT_FAM_MEMBERS',True)
my_string = ("Clients with family members of 2 are most numerous, followed by 1 (single persons),\n"
             "3 (families with one child) and 4. Clients with family size of 11 and 13 have 100% not \n"
             "repayment rate. Other families with 10 or 8 members have percents of not repayment of loans over 30%.\n"
             "Families with 6 or less members have repayment rates close to the 10% average.")
print(my_string)


plot_stats('NAME_INCOME_TYPE',False,False)
my_string = ("Most of applicants for loans are income from Working, followed by Commercial associate,\n"
             "Pensioner and State servant. The applicants with the type of income Maternity leave have \n"
             "almost 40% ratio of not returning loans, followed by Unemployed (37%). The rest of types of\n"
             "incomes are under the average of 10% for not returning loans.")
print(my_string)

plot_stats('OCCUPATION_TYPE',True, False)
my_string = ("Most of the loans are taken by Laborers, followed by Sales staff. IT staff take the lowest amount of loans.\n"
             "The category with highest percent of not repaid loans are Low-skill Laborers (above 17%), followed by Drivers\n"
             "and Waiters/barmen staff, Security staff, Laborers and Cooking staff.")
print(my_string)

plot_stats('ORGANIZATION_TYPE',True, False)
my_string = ("Oraganizations with highest percent of loans not repaid are Transport: type 3 (16%), Industry:\n"
             "type 13 (13.5%), Industry: type 8 (12.5%) and Restaurant (less than 12%).")
print(my_string)

plot_stats('NAME_EDUCATION_TYPE',True)
my_string = ("Majority of the clients have Secondary / secondary special education, followed by clients with Higher education.\n"
             "Only a very small number having an academic degree. The Lower secondary category, although rare, have the largest\n"
             "rate of not returning the loan (11%). The people with Academic degree have less than 2% not-repayment rate.")
print(my_string)

plot_stats('NAME_HOUSING_TYPE',True)
my_string = ("Over 250,000 applicants for credits registered their housing as House/apartment. \n"
             "Following categories have a very small number of clients (With parents, Municipal appartment).\n"
             "From these categories, Rented apartment and With parents have higher than 10% not-repayment rate.")
print(my_string)

# Plot distribution of one feature
def plot_distribution(feature,color):
    plt.figure(figsize=(10,6))
    plt.title("Distribution of %s" % feature)
    sns.histplot(df_train[feature].dropna(),color=color, kde=True,bins=100)
    plt.show()   
    
# Plot distribution of multiple features, with TARGET = 1/0 on the same graph
def plot_distribution_comp(var,nrow=2):
    
    i = 0
    t1 = df_train.loc[df_train['TARGET'] != 0]
    t0 = df_train.loc[df_train['TARGET'] == 0]

    sns.set_style('whitegrid')
    plt.figure()
    fig, ax = plt.subplots(nrow,2,figsize=(12,6*nrow))

    for feature in var:
        i += 1
        plt.subplot(nrow,2,i)
        sns.kdeplot(t1[feature], bw_adjust=1,label="TARGET = 1")
        sns.kdeplot(t0[feature], bw_adjust=1,label="TARGET = 0")
        plt.ylabel('Density plot', fontsize=12)
        plt.xlabel(feature, fontsize=12)
        locs, labels = plt.xticks()
        plt.tick_params(axis='both', which='major', labelsize=12)
    plt.show();
    
plot_distribution('AMT_INCOME_TOTAL','green')
plot_distribution('AMT_CREDIT','blue')
plot_distribution('AMT_ANNUITY','tomato')
plot_distribution('AMT_GOODS_PRICE','brown')
plot_distribution('DAYS_BIRTH','blue')
plot_distribution('DAYS_EMPLOYED','red')
plot_distribution('DAYS_REGISTRATION','green')
plot_distribution('DAYS_ID_PUBLISH','blue')

var = ['AMT_ANNUITY','AMT_GOODS_PRICE','DAYS_EMPLOYED', 'DAYS_REGISTRATION','DAYS_BIRTH','DAYS_ID_PUBLISH']
plot_distribution_comp(var,nrow=3)

plot_stats('REG_REGION_NOT_LIVE_REGION')
plot_stats('REG_REGION_NOT_WORK_REGION')
my_string=("Very few people are registered in not live or not work region. Generally, the rate of not return is \n"
           "slightly larger for these cases than in the rest (slightly above 8% compared with approx. 8%")
print(my_string)


plot_stats('REG_CITY_NOT_LIVE_CITY')
plot_stats('REG_CITY_NOT_WORK_CITY')
my_string = ("Generally, much more people register in the city they live or work (a larger number register differently\n"
             "in the working city than living city). The ones that register in different city than the working or living\n"
             "city are more frequently not-repaying the loans than the ones that register same city (work 11% or live 12%).")
print(my_string)


#  [Problem 5] (Advanced task) Posting to Notebooks 
# -------------------------------------------------
# Although not required, we encourage you to post your EDA notebook on the Notebooks page of the competition.
# The competition is expired!!!
```
